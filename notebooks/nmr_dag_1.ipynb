```python

import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib
from lmfit import models
import utils.scopeutil as scope
from notebooks.utils.OOMFormatter import OOMFormatter
```

```python
SI_PULSE_SIGNAL = 1
SI_READ_SIGNAL = 2

ds = {}
for delay in range(400, 850, 50):
    ds[delay] = {}
    for signal_index in [SI_PULSE_SIGNAL, SI_READ_SIGNAL]:
        df = scope.df_for_csv(scope.path_for_measurement_params(delay, signal_index))
        ds[delay][signal_index] = df
```

```python
for delay, signals in ds.items():
    fig, ax = plt.subplots()

    ax.plot(signals[SI_PULSE_SIGNAL]['x'], signals[SI_PULSE_SIGNAL]['y'])
    ax.plot(signals[SI_READ_SIGNAL]['x'], signals[SI_READ_SIGNAL]['y'])

    plt.title(f"Delay {delay}")

    plt.xlabel("t (s)")
    plt.ylabel("U (V)")

    ax.xaxis.set_major_formatter(OOMFormatter(-4, "%1.1f"))

    plt.show()
```
