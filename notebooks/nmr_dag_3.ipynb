```python
# import pandas as pd
# import numpy as np
# import math
import matplotlib.pyplot as plt
import matplotlib as mpl
# from lmfit import models
from utils.scopeutil import *
from utils.OOMFormatter import OOMFormatter

mpl.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots()

for (delay, df) in get_df_of_all_delays(
        stof=STOF_KOPERCHLORIDE_ACETOON,
        verhouding="1_1",
        T=T1,
        signal_index=SI_SIGNAL)[4:]:

    ax.plot(df['x'], df['y'], alpha=1)

plt.xlabel(r"Delay [sec]")
plt.ylabel(r"Gemeten spanning, evenredig met $M_y$ [$V$]")
ax.xaxis.set_major_formatter(OOMFormatter(-3, "%1.1f"))
plt.ylim([0, 40])
plt.title("Ruwe T1 metingen koperchloride aceton 1:1")
plt.savefig("../plots/raw_T1_koperchloride_acetoon_1_1.png", dpi=800)
fig.show()
```
